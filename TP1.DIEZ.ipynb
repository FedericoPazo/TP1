```python
#buscamos en title la cantidad de ambientes de los deptos para que traiga mayusculas y minusculas de amb, AMB, ambientes, dormitorios ,ect

pattern = r"\d\s\w*"

pattern_rooms_regex = re.compile(pattern , re.IGNORECASE)


resultado_rooms1= data.title.apply(lambda x: x if x is np.NaN else pattern_rooms_regex.search(x))

print(resultado_rooms1)
```

```python
#cuantos valores de ambientes puedo obtener
resultado_rooms1.value_counts().sum()
```

```python
# en que columnas encontre datos
mask_notnull_rooms1= resultado_rooms1.notnull()
mask_notnull_rooms1
```
